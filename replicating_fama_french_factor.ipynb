### Book Equity

Compute book equity (`be`) following the Fama-French methodology's post-2016 adjustments.

- **Preferred Stock (`ps`)** is constructed using a fallback hierarchy:
  - Use `pstkrv` (preferred stock redemption value) if available.
  - If missing, fall back to `pstkl` (preferred stock liquidating value).
  - If still missing, use `pstk` (preferred stock book/carrying value).
  - If all are missing, assign 0.

- **Book Equity (`be`)** is computed as:
  be = seq - ps
  where `seq` is stockholders’ equity.

Filter to keep only firms with **positive book equity**, since negative or zero BE implies severe financial distress or accounting anomalies

```python
# preferrerd stock
df_comp['ps']=np.where(df_comp['pstkrv'].isnull(), df_comp['pstkl'], df_comp['pstkrv']) # use pstkl if pstkrv is missing
df_comp['ps']=np.where(df_comp['ps'].isnull(),df_comp['pstk'], df_comp['ps']) # if ps is still missing, use pstk
df_comp['ps']=np.where(df_comp['ps'].isnull(),0,df_comp['ps']) # if ps is still missing, make it 0
```

```python
# create book equity 
df_comp['be'] = df_comp['seq'] - df_comp['ps'] # stockholder's Equity - Preferred share
df_comp['be'] = df_comp['be'].where(df_comp['be'] > 0, np.nan) # keep BE only if its positive
```

```python
# sort each firm's accounting records in chronological order
df_comp = df_comp.sort_values(by=['gvkey', 'datadate'])

# add a counter to track how many years of data exist per firm
df_comp['count'] = df_comp.groupby('gvkey').cumcount()

# retain only relevant columns for merging and future use
df_comp = df_comp[['gvkey', 'datadate', 'year', 'be', 'count']]
```

### Firm-Level Market Equity (`ME`) Aggregation

Since companies (`permco`) may issue multiple securities (`permno`), the replication aggregate market equity (`ME`) to the **firm level**:

- sum **ME** across all `permno` under a `permco` to get the firm’s total market cap.
- identify the **`permno`** with the largest ME and use it as firm's representative security.
- Finally, replace the security-level ME with the firm-level total ME.

```python
# sum of ME across all permnos belonging to the same permco on a given date
crsp_sumME = df_crsp.groupby(['jdate', 'permco'])['ME'].sum().reset_index()

# find the largest single ME (security) per permco/date
crsp_maxME = df_crsp.groupby(['jdate', 'permco'])['ME'].max().reset_index()

# keep only the permno with the largest ME per firm/date
df_crsp1 = pd.merge(df_crsp, crsp_maxME, how='inner', on=['jdate', 'permco', 'ME'])

# drop ME (security-level) and replace it with firm-level total ME
df_crsp1 = df_crsp1.drop(['ME'], axis=1)

# add back the total firm-level ME
df_crsp2 = pd.merge(df_crsp1, crsp_sumME, how='inner', on=['jdate', 'permco'])  # New 'ME' is total ME

# sort and drop duplicates (in case of multiple securities with identical ME)
df_crsp2 = df_crsp2.sort_values(by=['permno', 'jdate']).drop_duplicates()

df_crsp2.head()
```

### Fama-French Fiscal Year and Lagged Size

- Extract December ME **`dec_ME`** to use as lagged size in the following June’s portfolio formation.
- Fama-French portfolios run from **July to June**, so need to construct a custom fiscal year: **`ffyear`, `ffmonth`**
- A **`(1 + retx)`** column for cumulative return calculations, then sort by **`permno`** and **date** to enable time series operations.

```python
# extract year and month from jdate (used later for matching December and computing FF fiscal periods)
df_crsp2['year'] = df_crsp2['jdate'].dt.year
df_crsp2['month'] = df_crsp2['jdate'].dt.month

# keep December market equity to use as lagged size in next June's portfolio formation
decme = df_crsp2[df_crsp2['month'] == 12]
decme = decme[['permno', 'date', 'jdate', 'ME', 'year']].rename(columns={'ME': 'dec_ME'})
decme
```

```python
# create Fama-French fiscal month (July to June)
df_crsp2['ffdate'] = df_crsp2['jdate'] + MonthEnd(-6)  # subtract 6 months to make July the start
df_crsp2['ffyear'] = df_crsp2['ffdate'].dt.year        # ffyear corresponds to June-end
df_crsp2['ffmonth'] = df_crsp2['ffdate'].dt.month
```

```python
# prepare for cumulative return calculation
df_crsp2['1+retx'] = 1 + df_crsp2['retx']

# ensure CRSP data is sorted by stock and date for time-series operations
df_crsp2 = df_crsp2.sort_values(by=['permno', 'date'])
df_crsp2
```

### Lagged Market Cap and Return Weights for Portfolio Construction

To compute value-weighted returns for Fama-French portfolios, need to assign **weights** to each stock each month. Fama and French use **lagged market equity** for this purpose. 

- **Cumulative Return (`cumretx`)**: Calculate the cumulative return over the fiscal year (July to June) for each stock.
- **Lagged Cumulative Return (`lcumretx`)**: Shift the cumulative return by one month to avoid look-ahead bias.
- **Lagged Market Equity (`lme`)**: Market equity is lagged by one month to match returns with information available at the start of the period.
  - For the first observation of each stock, backfill `lme` as:  
    lme = ME / (1 + retx)
- **Baseline ME (`mebase`)**: For each fiscal year, store January’s lagged ME. This acts as the reference point for calculating weights for the rest of the year.
- **Final Weight (`wt`)**:
  - For January (`ffmonth == 1`), the weight is just the lagged market equity (`lme`).
  - For other months, compute the updated ME using January's `mebase` scaled by cumulative return:
    wt = mebase * lcumretx
- **December ME for Size Sorting**: Increment the year on `dec_me` to align it with **June of the following year**, which is when size breakpoints are applied.
- **June Snapshot (`df_crsp_jun`)**: Filter only June records and merge with December ME (`dec_me`) to use for **size & value sorting** into portfolios.

```python
# cumulative return over fiscal year (July–June) for each stock
df_crsp2['cumretx'] = df_crsp2.groupby(['permno', 'ffyear'])['1+retx'].cumprod()

# lag the cumulative return by one month to avoid look-ahead bias
df_crsp2['lcumretx'] = df_crsp2.groupby('permno')['cumretx'].shift(1)

# lag market equity (ME) by one month
df_crsp2['lme'] = df_crsp2.groupby('permno')['ME'].shift(1)

# for the first appearance of a stock, back-calculate lagged ME using ME / (1 + retx)
df_crsp2['count'] = df_crsp2.groupby('permno').cumcount()
df_crsp2['lme'] = np.where(df_crsp2['count'] == 0,
                           df_crsp2['ME'] / df_crsp2['1+retx'],
                           df_crsp2['lme'])
```

```python
# for each permno and ffyear, get January’s lagged ME to use as baseline
df_mebase = df_crsp2[df_crsp2['ffmonth'] == 1][['permno', 'ffyear', 'lme']]
df_mebase = df_mebase.rename(columns={'lme': 'mebase'})

# merge baseline ME back into main dataset
df_crsp3 = pd.merge(df_crsp2, df_mebase, how='left', on=['permno', 'ffyear'])

# compute final return weight:
# - For January, use lme directly
# - For other months, scale mebase using cumulative returns
df_crsp3['wt'] = np.where(df_crsp3['ffmonth'] == 1,
                          df_crsp3['lme'],
                          df_crsp3['mebase'] * df_crsp3['lcumretx'])
```

```python
# increment year to match next June
df_decme = decme.copy()
df_decme['year'] = df_decme['year'] + 1
df_decme = df_decme[['permno', 'year', 'dec_ME']]
df_decme
```

```python
# filter only June observations
df_crsp3_jun = df_crsp3[df_crsp3['month'] == 6] # Fama & French use June of each year to sort stocks based on size and Boot-to-Market (B/M)

# merge with lagged December ME
df_crsp_jun = pd.merge(df_crsp3_jun, df_decme, how='inner', on=['permno','year'])

# keep only relevant columns
df_crsp_jun = df_crsp_jun[[
    'permno', 'date', 'jdate', 'shrcd', 'exchcd',
    'ret', 'ME', 'wt', 'cumretx', 'mebase', 'lme', 'dec_ME', 'ticker', 'cusip'
]]
# drop duplicates if any
df_crsp_jun = df_crsp_jun.sort_values(by=['permno', 'jdate']).drop_duplicates()
df_crsp_jun
```

### CRSP/Compustat Link Table

**CRSP/Compustat Merged Link Table** is used to connect company-level data from **Compustat** with security-level data from **CRSP**

- **gvkey**: Unique firm identifier in Compustat.
- **permno**: Unique security identifier in CRSP (renamed from `lpermno` in the link table).
- **linktype**: Type of link between Compustat and CRSP. Filter these with values beginning with `'L'`, which indicate valid, active links.
- **linkprim**: Indicates whether the link is primary.
  - `'C'`: Primary Compustat link (most reliable).
  - `'P'`: Primary CRSP link (used if `'C'` is not available).
- **linkdt**: Start date of the valid link between Compustat firm (`gvkey`) and CRSP security (`permno`).
- **linkenddt**: End date of the valid link. If missing, it means the link is still active — set it to today’s date to avoid excluding ongoing links.

```python
# CCM Block
df_ccm = conn.raw_sql("""
    SELECT gvkey, lpermno AS permno, linktype, linkprim, 
           linkdt, linkenddt
    FROM crsp.ccmxpf_linktable
    WHERE SUBSTR(linktype, 1, 1) = 'L'       -- Only active "link" types (e.g., LU, LC)
      AND (linkprim = 'C' OR linkprim = 'P') -- Primary links: 'C' (most reliable), 'P' (used if 'C' missing)
""", date_cols=['linkdt', 'linkenddt'])
#df_ccm = pd.read_csv("ccm.csv")
df_ccm.head()
```

### Portfolio Formation

To construct the Fama-French value (HML) and size (SMB) portfolios, assign each firm to one of six portfolios based on their size (market equity) and book-to-market (B/M) ratio.

#### Step 1: NYSE Breakpoints
- Use only NYSE stocks to compute **breakpoints** for portfolio assignment.
- **Size**: NYSE median ME per `jdate` is used to divide firms into:
  - `'S'` (small) if ME ≤ median
  - `'B'` (big) otherwise
- **Value**: NYSE B/M breakpoints (30th and 70th percentiles) split firms into:
  - `'Low'` (low B/M), `'Medium'` (medium B/M), `'High'` (high B/M)

#### Step 2: Assign Firms to Portfolios
- All stocks (not just NYSE) are assigned to one of 6 portfolios using the NYSE breakpoints:
  - **S/L**, **S/M**, **S/H**
  - **B/L**, **B/M**, **B/H**

#### Step 3: Filter Valid Firms
- Keep only firms that:
  - Have **at least 2 years of Compustat data**
  - Have **positive book equity and ME**
  - Are **common stocks (shrcd = 10 or 11)**
  - Have been successfully assigned a portfolio (`szport` and `bmport`)

#### Step 4: Store June Assignments
- June assignments are stored with a `ffyear` flag, indicating the **year of portfolio formation**.
- These assignments are used to merge with monthly CRSP returns from **July of that year to June of the following year**.

```python
# compute the NYSE median market equity (ME) for each portfolio formation date (jdate)
df_nyse_sz = df_nyse.groupby('jdate')['ME'].median().reset_index().rename(columns={'ME': 'sizemedn'})

# compute 30th and 70th percentiles for book-to-market (beme) by jdate
df_nyse_bm = df_nyse.groupby('jdate')['beme'].quantile([0.3, 0.7]).unstack().reset_index()
df_nyse_bm = df_nyse_bm.rename(columns={0.3: 'bm30', 0.7: 'bm70'})

# merge size and book-to-market breakpoints into a single dataframe
df_nyse_breaks = pd.merge(df_nyse_sz, df_nyse_bm, on='jdate', how='inner')

# merge NYSE-based breakpoints back with the full CRSP+Compustat universe (not just NYSE)
df_ccm1_jun = pd.merge(df_ccm_jun, df_nyse_breaks, on='jdate', how='left')
df_ccm1_jun
```

```python
# define a function to assign size portfolio ('S' for small, 'B' for big)
def sz_asgn(row):
    if pd.isnull(row['ME']) or pd.isnull(row['sizemedn']):
        return ''
    elif row['ME'] <= row['sizemedn']:
        return 'S'
    else:
        return 'B'

# define a function to assign value portfolio based on B/M:
# 'L' for Low, 'M' for Medium, 'H' for High
def bm_cls(row):
    if pd.isnull(row['beme']):
        return 'Missing B/M ratio'
    elif pd.isnull(row['bm30']) or pd.isnull(row['bm70']):
        return 'Missing NYSE breakpoints'
    elif row['count'] < 1:
        return 'Insufficient Compustat history'
    else:
        if row['beme'] <= row['bm30']:
            return 'Low'
        elif row['beme'] <= row['bm70']:
            return 'Medium'
        else:
            return 'High'
```

```python
# define valid rows to ensure clean application of bucket functions
valid_rows_sz = (df_ccm1_jun['ME'].notnull()) & (df_ccm1_jun['sizemedn'].notnull()) & (df_ccm1_jun['count'] >= 1)
valid_rows_bm = (df_ccm1_jun['beme'].notnull()) & (df_ccm1_jun['bm30'].notnull()) & (df_ccm1_jun['bm70'].notnull()) & (df_ccm1_jun['count'] >= 1)

# apply size bucket assignment
df_ccm1_jun.loc[valid_rows_sz, 'szasgn'] = df_ccm1_jun.loc[valid_rows_sz].apply(sz_asgn, axis=1)
df_ccm1_jun.loc[~valid_rows_sz, 'szasgn'] = ''  # mark as missing if not valid

# apply book-to-market bucket assignment
df_ccm1_jun['bmcls'] = df_ccm1_jun.apply(bm_cls, axis=1)

# separate remarks column for non-assigned classifications
df_ccm1_jun['remarks'] = df_ccm1_jun['bmcls'].where(~df_ccm1_jun['bmcls'].isin(['Low', 'Medium', 'High']), '')

# reset bmcls to ''
df_ccm1_jun['bmcls'] = df_ccm1_jun['bmcls'].where(df_ccm1_jun['bmcls'].isin(['Low', 'Medium', 'High']), '')

df_ccm1_jun
```

### Fama-French HML Factor Construction

The High Minus Low (HML) factor captures the return spread between **value stocks** (high book-to-market) and **growth stocks** (low book-to-market).

#### Steps:

1. **Value-Weighted Returns**:
   - For each month (`jdate`), compute value-weighted returns of the 6 portfolios formed by:
     - **Size**: Small (S) or Big (B)
     - **Value**: Low (L), Medium (M), High (H) Book-to-Market

2. **Portfolio Combination**:
   - Combine the **SH** and **BH** portfolios to represent High B/M stocks
   - Combine the **SL** and **BL** portfolios to represent Low B/M stocks

3. **HML Calculation**:
   - Compute the average return of High B/M stocks (`WH`)
   - Compute the average return of Low B/M stocks (`WL`)
   - **HML = WH - WL**

```python
# this function computes value-weighted returns for each size (S/B) × value (L/M/H) portfolio
def wavg(group, avg_name, weight_name):
    d = group[avg_name]
    w = group[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return np.nan
```

```python
# compute value-weighted average returns for each S/B and L/M/H portfolio
df_vwret = df_ccm4.groupby(['jdate', 'szasgn', 'bmcls']).apply(wavg, 'ret', 'wt').to_frame().reset_index()
df_vwret = df_vwret.rename(columns={0: 'vwret'})

# combine size and value labels into single portfolio code
df_vwret['szbm'] = df_vwret['szasgn'] + '_' + df_vwret['bmcls']

# pivot data: one column per S/B portfolio return
df_ff_hml = df_vwret.pivot(index='jdate', columns='szbm', values='vwret').reset_index()
```

```python
# compute HML factor:
# WH = average return of High B/M stocks (SH + BH) / 2
# WL = average return of Low B/M stocks (SL + BL) / 2
# HML = WH - WL
df_ff_hml['WH'] = (df_ff_hml['S_High'] + df_ff_hml['B_High']) / 2
df_ff_hml['WL'] = (df_ff_hml['S_Low'] + df_ff_hml['B_Low']) / 2
df_ff_hml['HML'] = df_ff_hml['WH'] - df_ff_hml['WL']

# compute SMB factor
df_ff_hml['WB']=(df_ff_hml['B_Low']+df_ff_hml['B_Medium']+df_ff_hml['B_High'])/3
df_ff_hml['WS']=(df_ff_hml['S_Low']+df_ff_hml['S_Medium']+df_ff_hml['S_High'])/3
df_ff_hml['SMB'] = df_ff_hml['WS']-df_ff_hml['WB']
```

```python
# rename jdate column to date for final output
df_ff_hml = df_ff_hml.rename(columns={'jdate': 'date'})
df_ff_hml['HML'] = df_ff_hml['HML'] * 100
df_ff_hml['SMB'] = df_ff_hml['SMB'] * 100
df_ff_hml[['date', 'HML', 'SMB']]
```

```python
# remove rows with nan (First few rows have nan because I calculated HML for firms with atleast 2 years of compustat)
df_ff_hml = df_ff_hml[df_ff_hml['HML'].notna()]
```

```python
# function to compute correlation for a given factor in a date range
def factor_corr_in_range(df, start_year, end_year, factor):
    df_filtered = df[
        (df['date'].dt.year >= start_year) & 
        (df['date'].dt.year <= end_year)
    ]
    corr = df_filtered[[f'my_{factor}', f'FF_{factor}']].corr().iloc[0, 1]
    return round(corr, 4)

# time periods to evaluate
time_periods = [
    (2022, 2024),
    (2016, 2024),
    (2001, 2024),
    (1962, 2024)
]

# compute and display correlations
for start, end in time_periods:
    hml_corr = factor_corr_in_range(df_hml_smb_compare, start, end, 'HML')
    smb_corr = factor_corr_in_range(df_hml_smb_compare, start, end, 'SMB')
    print(f"Correlation ({start}–{end}): HML = {hml_corr}, SMB = {smb_corr}")
```

The 2022–2024 correlations of 0.9977 for HML and 0.9976 for SMB indicate that this replication captures 99.77% and 99.76% of the variation in the official Fama-French HML and SMB factors, respectively. However, as we extend the comparison to longer periods, this correlation gradually decreases. This is because the Fama-French used additional proprietary and historical datasets such as Moody's Manuals and hand-collected book equity values, and have refined their methodology over time. For instance, earlier versions of Fama-French book equity explicitely includes deferred taxes and investment tax credits, but their updated methodology (following FASB 109) excludes these. I followed the latest approach for entire period and did not rely on delisting returns.

### Why My HML & SMB Differ Slightly from Fama-French Factors?


#### Methodological Differences

- **2-Year Compustat History Requirement**  
  In the replication, required firms to have **at least 2 years of Compustat data** (`count >= 1`) to ensure sufficient accounting history.  
  >While the Fama-French 2023 paper mentions the importance of clean and accurate Compustat data for computing book equity, it does **not explicitly specify 2 years** restriction.

- **Missing Book Equity Filtering**  
  The replication explicitly exclude firms with **non-positive book equity (`be <= 0`)**.  
  > Fama and French mentions that they constructed book-to-market ratios only where meaningful, but could have dealt cases differently.

#### Linking and Timing Issues

- **CRSP-Compustat Linking**  
  The replication used the **WRDS-recommended filters** (`linktype LIKE 'L%'` and `linkprim in ('C','P')`) and enforced **link date bounds (`jdate` ∈ [linkdt, linkenddt])**.  
  > Fama-French mention, they only started using CRSP-Compustat linking after 2021. And do not provide much information on filters.

#### Data Differences

- **Raw vs. Adjusted CRSP Returns**  
  Used ret (monthly returns including dividends, but not necessarily delisting), but
    > FF incorporated **data corrections or exclusions** for properly handeling delisting that aren't present in the raw WRDS download.

#### Calculation Choices

- **Handling of Ties and Missing Weights**  
  FF's production code might treat **ties in market equity**, **zero-weighted stocks**, or **missing returns** slightly differently.

- **Precision and Rounding**  
  Fama-French publish monthly HML values rounded to **two decimal places**, while this replication calculations preserve higher precision.

### Summary

> Even with careful alignment, exact replication of the Fama-French HML factor is difficult due to **proprietary adjustments**, **hand-collected inputs**, and **minor methodological choices**.  
> However, the replicated HML matches closely and is appropriate for empirical asset pricing work.
